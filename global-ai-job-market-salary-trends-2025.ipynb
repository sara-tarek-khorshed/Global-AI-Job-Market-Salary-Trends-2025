```python
plt.figure(figsize=(12 ,8))
sns.histplot( df["job_title"] , bins=20 ,kde=True , color="blue" )
plt.xticks(rotation=90)
plt.title("job_title" ,fontsize=26)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.boxplot(x="company_size", y="salary_usd", data=df )
plt.title("Salary Distribution by Company Size")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(df ,x="company_size" ,y="salary_usd" ,estimator=sum )
plt.xlabel("company_size")
plt.ylabel("salary_usd")
plt.title("Company Size vs. Total Salary Paid" ,fontsize=20)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(df ,x="experience_level" ,y="salary_usd" ,estimator=sum )
plt.title("Experience Level vs. Total Salary Paid" ,fontsize=20)
```

```python
fig=px.bar(df ,x="job_title" ,y="salary_usd" , hover_data=["experience_level" ,"employment_type" ,"remote_ratio"] , color="years_experience")
fig.show()#remote_ratio       years_experience
```

```python
fig=px.bar(df ,x="company_size" ,y="salary_usd" )
fig.show()
```

```python
all_reqs = df["required_skills"].dropna().str.split(",").explode()
all_reqs = all_reqs.str.strip() 

req_counts = all_reqs.value_counts()

plt.figure(figsize=(12,6))
req_counts.plot(kind='bar', color='red')
plt.title(" Common Job Requirements", fontsize=16)
plt.xlabel("Requirement")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
all_reqs = df["required_skills"].dropna().str.split(",").explode()
all_reqs = all_reqs.str.strip() 

req_counts = all_reqs.value_counts()

plt.figure(figsize=(12,6))
req_counts.head(10).plot(kind='bar', color='skyblue')
plt.title("Top 10 Most Common Job Requirements", fontsize=16)
plt.xlabel("Requirement")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
industry_counts=df["industry"].value_counts()
plt.figure(figsize=(10,8))
industry_counts.plot(kind="bar" ,color="skyblue")
plt.title(" Common Industries" ,fontsize=20)
plt.xlabel("Industry")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()






```
